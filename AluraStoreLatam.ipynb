#1. Análisis de facturación



```python
# Calcular ingresos por tienda
ingreso_tienda1 = tienda1["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()
```

```python
#Mostrando ingreso total de cada tienda
print("El ingreso total de la tienda 1 es: ",ingreso_tienda1)
print("El ingreso total de la tienda 2 es: ",ingreso_tienda2)
print("El ingreso total de la tienda 3 es: ",ingreso_tienda3)
print("El ingreso total de la tienda 4 es: ",ingreso_tienda4)
```

```python
# Gráfico de barras: Ingresos por tienda
ingresos = [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]
tiendas = ["tienda 1", "tienda 2", "tienda 3", "tienda 4"]

plt.figure(figsize=(5,3))
plt.bar(tiendas, ingresos, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.title("Ingresos por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingresos totales")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 2. Ventas por categoría

```python
# Cantidad de productos vendidos por categoría en cada tienda

ventas_categoria_t1 = tienda1.groupby("Categoría del Producto")["Producto"].count()
ventas_categoria_t2 = tienda2.groupby("Categoría del Producto")["Producto"].count()
ventas_categoria_t3 = tienda3.groupby("Categoría del Producto")["Producto"].count()
ventas_categoria_t4 = tienda4.groupby("Categoría del Producto")["Producto"].count()
```

```python
print("Tienda 1 - Ventas por categoría")
display(ventas_categoria_t1.sort_values(ascending=False))

print("Tienda 2 - Ventas por categoría")
display(ventas_categoria_t2.sort_values(ascending=False))

print("Tienda 3 - Ventas por categoría")
display(ventas_categoria_t3.sort_values(ascending=False))

print("Tienda 4 - Ventas por categoría")
display(ventas_categoria_t4.sort_values(ascending=False))
```

```python
import numpy as np
ventas_por_categoria = pd.DataFrame({
    'Tienda 1': ventas_categoria_t1,
    'Tienda 2': ventas_categoria_t2,
    'Tienda 3': ventas_categoria_t3,
    'Tienda 4': ventas_categoria_t4
})

# Usar un estilo de Matplotlib para mejorar la estética
plt.style.use('seaborn-v0_8-darkgrid')

fig, ax = plt.subplots(figsize=(8, 5))

categories = ventas_por_categoria.index
bar_width = 0.2

# Definir posiciones para las barras agrupadas
r1 = np.arange(len(categories))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]
r4 = [x + bar_width for x in r3]

# Crear las barras para cada tienda
ax.bar(r1, ventas_por_categoria['Tienda 1'], color='skyblue', width=bar_width, edgecolor='white', label='Tienda 1')
ax.bar(r2, ventas_por_categoria['Tienda 2'], color='lightcoral', width=bar_width, edgecolor='white', label='Tienda 2')
ax.bar(r3, ventas_por_categoria['Tienda 3'], color='lightgreen', width=bar_width, edgecolor='white', label='Tienda 3')
ax.bar(r4, ventas_por_categoria['Tienda 4'], color='gold', width=bar_width, edgecolor='white', label='Tienda 4')

# Añadir etiquetas, título y leyenda
ax.set_xlabel('Categoría del Producto', fontsize=12)
ax.set_ylabel('Cantidad Vendida', fontsize=12)
ax.set_title('Ventas por Categoría en Todas las Tiendas', fontsize=16)
ax.set_xticks([r + 1.5 * bar_width for r in range(len(categories))])
ax.set_xticklabels(categories, rotation=45, ha='right')
ax.legend(title='Tienda', bbox_to_anchor=(1.05, 1), loc='upper left')

# Asegurar que las líneas de los ejes sean visibles y más prominentes
ax.spines['top'].set_visible(False) # Ocultar el eje superior
ax.spines['right'].set_visible(False) # Ocultar el eje derecho
ax.spines['left'].set_linewidth(1.5) # Hacer el eje izquierdo más grueso
ax.spines['bottom'].set_linewidth(1.5) # Hacer el eje inferior más grueso
ax.spines['left'].set_color('black') # Color del eje izquierdo
ax.spines['bottom'].set_color('black') # Color del eje inferior

plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Calificación promedio de cada tienda

promedio_t1 = tienda1["Calificación"].mean()
promedio_t2 = tienda2["Calificación"].mean()
promedio_t3 = tienda3["Calificación"].mean()
promedio_t4 = tienda4["Calificación"].mean()
```

```python
tienda1 ["Calificación"]
```

```python
print("Calificación promedio de cada tienda:")
print(f"Tienda 1: {promedio_t1:.2f}")
print(f"Tienda 2: {promedio_t2:.2f}")
print(f"Tienda 3: {promedio_t3:.2f}")
print(f"Tienda 4: {promedio_t4:.2f}")
```

```python
#Grafico de calificacion promedio por tienda

calificaciones_promedio = [promedio_t1, promedio_t2, promedio_t3, promedio_t4]

fig, ax = plt.subplots(figsize=(5, 3))

ax.bar(tiendas, calificaciones_promedio, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
ax.set_title('Calificación Promedio por Tienda', fontsize=15)
ax.set_xlabel('Tiendas', fontsize=10)
ax.set_ylabel('Calificación Promedio', fontsize=12)
ax.set_ylim(0, 5) # Las calificaciones van de 1 a 5

# Añadir el valor de la calificación en la parte superior de cada barra
for i, v in enumerate(calificaciones_promedio):
    ax.text(i, v + 0.1, f'{v:.2f}', ha='center', va='bottom', fontsize=10)

# Asegurar que las líneas de los ejes sean visibles y más prominentes
ax.spines['left'].set_linewidth(1.5) # Hacer el eje izquierdo más grueso
ax.spines['bottom'].set_linewidth(1.5) # Hacer el eje inferior más grueso
ax.spines['left'].set_color('black') # Color del eje izquierdo
ax.spines['bottom'].set_color('black') # Color del eje inferior

ax.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
